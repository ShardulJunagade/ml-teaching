```python
from sklearn.datasets import make_moons

X, y = make_moons(n_samples = 1000, noise = 0.1, random_state = 0)

plt.figure(figsize = (8, 6))
plt.scatter(X[:, 0], X[:, 1], c = y, s = 40, cmap = plt.cm.Spectral);
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 100, random_state = 0)
rf.fit(X, y)
```

```python
# Decision surface
plt.figure(figsize = (8, 6))
plt.scatter(X[:, 0], X[:, 1], c = y, s = 40, cmap = plt.cm.viridis)
ax = plt.gca()
xlim = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
ylim = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5

# Create grid to evaluate model
x_lin = np.linspace(xlim[0], xlim[1], 30)
y_lin = np.linspace(ylim[0], ylim[1], 30)

XX, YY = np.meshgrid(x_lin, y_lin)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = rf.predict_proba(xy)[:, 1].reshape(XX.shape)

# Plot decision boundary
ax.contourf(XX, YY, Z, cmap = plt.cm.viridis, alpha = 0.2)
plt.colorbar()
plt.show()
```

### Let's see some Plots and their Contours

* $$ \color{red}{F(x, y) = \sin(x)\cos(y)}, -\frac{3\pi}{2} \le x \le \frac{3\pi}{2}, -\pi \le y \le \pi $$
* $$ \color{blue}{G(x, y) = e^{-(x^2 + y^2)} \sin(x)}, -2 \le x \le 2, -2 \le y \le 2 $$
* $$ \color{green}{H(x,y) = \begin{cases} 1 & x^2 + y^2 < 1 \\ 0 & x^2 + y^2 \ge 1 \end{cases}}, -3 \le x \le 3, -3 \le y \le 3 $$

```python
from mpl_toolkits.mplot3d import Axes3D

# First function F
X1 = np.linspace(-3/2 * np.pi, 3/2 * np.pi, 100)
Y1 = np.linspace(-np.pi, np.pi, 100)
XX1, YY1 = np.meshgrid(X1, Y1)
F = np.sin(XX1) * np.cos(YY1)

# Second function G
X2 = np.linspace(-3/2 * np.pi, 3/2 * np.pi, 100)
Y2 = np.linspace(-2, 2, 100)
XX2, YY2 = np.meshgrid(X2, Y2)
G = np.exp(-XX2**2 - YY2**2) * np.sin(XX2)

# Third function H
X3 = np.linspace(-3, 3, 100)
Y3 = np.linspace(-3, 3, 100)
XX3, YY3 = np.meshgrid(X3, Y3)
H = ((XX3**2 + YY3**2) < 1).astype(float)


fig = plt.figure(figsize=(16, 12))
# Subplot 1: Level surface of F
ax1 = fig.add_subplot(331)
contour1 = ax1.contourf(XX1, YY1, F, cmap = "inferno", levels = 10)
fig.colorbar(contour1, ax = ax1)
ax1.set_title(r"Contour Plot of $F(x, y) = \sin(x)\cos(y)$", fontsize = 14)

# Subplot 2: Level surface of G
ax2 = fig.add_subplot(334)
contour2 = ax2.contourf(XX2, YY2, G, cmap = "viridis", levels = 10)
fig.colorbar(contour2, ax = ax2)
ax2.set_title(r"Contour Plot of $G(x, y) = e^{-(x^2 + y^2)} \sin(x)$", fontsize = 14)

# Subplot 3: Level surface of H
ax3 = fig.add_subplot(337)
contour3 = ax3.contourf(XX3, YY3, H, cmap = "binary")
fig.colorbar(contour3, ax = ax3)
ax3.set_title(r"Contour Plot of $H(x, y) = x^2 + y^2 < 1$", fontsize = 14)

# Subplot 4: 3D graph of F
ax4 = fig.add_subplot(132, projection = "3d")
ax4.plot_surface(XX1, YY1, F, cmap = "inferno")
ax4.set_title(r"Graph of $F(x, y) = \sin(x)\cos(y)$", fontsize = 14)

# Subplot 5: 3D graph of G
ax5 = fig.add_subplot(133, projection = "3d")
ax5.plot_surface(XX2, YY2, G, cmap = 'viridis')
ax5.set_title(r"Graph of $G(x, y) = e^{-(x^2 + y^2)} \sin(x)$", fontsize = 14)

plt.tight_layout()
plt.show()
```

```python
from einops import rearrange, repeat, reduce
```

Usually when dealing with images, the shapes are of the form $(B \times C \times H \times W)$ where <br>
$B$: Batch Size <br>
$C$: Number of Image Channels <br>
$H$: Image Height <br>
$W$: Image Width <br>
Inorder to flatten the image data into a linear array $(B \times C \times H \times W) \to (B \times C \cdot H \cdot W)$, we may use <br>
$$ \texttt{einops.rearrange(img, "b c h w -> b (c h w)")} $$
